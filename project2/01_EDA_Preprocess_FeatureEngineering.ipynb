### Defining functions

```python
# Consider converting the corresponding months sold into season as research has reported that seasons does affect 
# sale price of houses.
def convert_month_season(months):
    """
    This function converts the corresponding months of the year into corresponding US seasons
    based on a defined dictionary below.
    
    - Summer: June - August
    - Fall: September - November
    - Winter: December - February
    - Spring: March - May 
    
    Takes in the following argument:
    @months: Integer value
    """
    month_to_season_dict = {
        1 : "Winter",
        2 : "Winter",
        3: "Spring",
        4: "Spring",
        5: "Spring",
        6: "Summer",
        7: "Summer",
        8: "Summer",
        9: "Fall",
        10: "Fall",
        11: "Fall",
        12: "Winter"
    }
    
    return month_to_season_dict[months]

def extract_col_with_less_unique_values(dataframe,unique_value_condition):
    """
    This function returns a list of columns from a dataframe which contains unique values less than or equals to the
    number of unique values.
    Takes in the following argument:
    @dataframe: pandas dataframe
    @unique_value_condition: Integer specifying the number of unique values as condition
    """
    column_data_with_limit_unique_values = []
    for column in dataframe.columns:
        if len(dataframe[column].value_counts(dropna = False)) <= unique_value_condition:
            column_data_with_limit_unique_values.append(column)
    return column_data_with_limit_unique_values

def barplot_annotation(plots,dataframe_entries):
    """
    This function takes in plots and number of dataframe entries which annotates
    the number of instance for each patches(e.g bars for bar charts) in plot, as well as the percentage of instances in relation to
    number dataframe entries.
    Takes in the following argument:
    @plots: plots by matplotlib that was generated from subplot_histogram function below.
    @dataframe_entries: number of dataframe entries which plots are made on.
    """
    for p in plots.patches:
        #Format the annotation, 4 decimals for percentage and no decimal for barplot values.
        plots.annotate(f"{p.get_height():.0f}\n({100*(p.get_height()/dataframe_entries):,.1f}%)", 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha = 'center', va = 'center', 
                       size=10,
                       xytext = (0, 20), 
                       textcoords = 'offset points')

def subplot_histograms(dataframe, list_of_columns, list_of_xlabels):
    import random
    from matplotlib.ticker import FuncFormatter, MaxNLocator
    """
    This function generates a n by 1 subplots of histogram for required columns of the dataframe.
    Takes in the following argument:
    @dataframe: pandas dataframe
    @list_of_columns: names of corresponding pandas dataframe columns
    @list_of_xlabels: list of xlabels for use in labelling the x-axis of histogram plots
    """
    if len(list_of_columns) > 1:
    # Makes sure you have enough rows
        fig, ax = plt.subplots(nrows=len(list_of_columns), ncols=1,figsize = (15,10*len(list_of_columns))) # You'll want to specify your figsize
        for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
            sns.barplot(x = dataframe[column].value_counts(dropna = False).index, y = dataframe[column].value_counts(), ax = ax[i])

            # Set titles, labels, etc here for each subplot
            ax[i].set_title("Histogram plot for column {}".format(list_of_columns[i]),fontsize = 15)
            ax[i].set_xlabel(list_of_xlabels[i],fontsize = 15)
            ax[i].set_ylabel("Counts",fontsize = 15)

            #Annotate the plot's patches
            barplot_annotation(ax[i],len(dataframe))
    else:
        fig, ax = plt.subplots(figsize = (15,20)) # You'll want to specify your figsize
        single_col = list_of_columns[0]
        ax = dataframe[single_col].value_counts(dropna = False).plot(kind='bar')
        # Set titles, labels, etc here for each subplot
        ax.set_title("Histogram plot for column {}".format(single_col),fontsize = 15)
        ax.set_xlabel(list_of_xlabels[0],fontsize = 15)
        ax.set_ylabel("Counts",fontsize = 15)

        #Annotate the plot's patches
        barplot_annotation(ax,len(dataframe))

    #Prevents overlap of titles,axes
    plt.tight_layout()
    return None
```

## Conclusions
Lasso has by far has the best test RMSE value at 33407. With 40 features made up of 24 numerical features and 16 categorical features expanded from 3 original categorical features, it is now more ideal and easy to explain to relevant stakeholders. Ideally, we would always prefer to have a model that is complex enough to fit all the points but yet could simply be explained.

Additionally, the Lasso regression model does not look to be overfitted given that it has a slightly higher training RMSE value compared to Ridge regression as shown.

## Top 3 features from model effecting property price
Looking at the absolute values of the coefficients from the Lasso Model, we see that the heating quality and condition plays a key feature in determining the house price with a coefficient of 25223. The next few key features would be total porch area and wood deck area with coefficients of -21617 and 13352 respectively.

It is no doubt that heating quality and condition plays a significant part for Iowa citizens as it could be as cold as -10°C. Furthermore, given the state history of encountering floods in the past which are caused by rapid snow melt and heavy summer rainstorms, it is undeniable that Iowa citizens would place more emphasis on heating quatlity and condition for their home. Ref: https://www.britannica.com/place/Iowa-state/Climate.

It is interesting to note that larger porch area actually reduces the property price. A possible explanation could be due to its past experiences with natural disasters which tends to destroy or damage house property, hence causing damages to porch which costs on average USD10000 to USD20000 for rebuilding, thus widely frowned upon by buyers(ref: https://www.fixr.com/costs/porch-addition).

At number three, the wood deck area of the property takes the spot and is no surprise that wood decks helps provides comfort & is heat-resistant which is good during the summer days which are long, warm, humid, which could be a possible reason why it is preferred 

Therefore, we can conclude that the heating quality and condition, total porch area and wood deck area are key features in affecting the property price

## Recommendations
The above conclusions are purely based on the Ames housing dataset with some imputation made. In order to complement the above findings, more granular data such as buyer data, seller financial background and coordinates of home sales would be beneficial. This would facilitate more studies such as buyer and seller behaviour or neighbourhood studies which in turn, would enable home owners or property developers to better target buyers, or to even educate buyers on the type of houses based on subsequent findings from potential studies.
